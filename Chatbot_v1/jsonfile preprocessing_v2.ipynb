```python
import json
import pandas as pd
```

```python
with open('./data/speech_data.json', 'r', encoding = 'utf-8') as json_file:
    json_data = json.loads(json_file.read())
```

```python
sentence_list = [list(json_data[i]['talk']['content'].values()) for i in range(len(json_data))]
```

```python
a = 0
b = 1
person_1 = []
person_2 = []
for i in range(len(sentence_list)):
    for ii in range(3):
        person_1.append(sentence_list[i][a])
        person_2.append(sentence_list[i][b])
        a += 2
        b += 2
        if a == 6:
            a = 0
        if b == 7:
            b = 1
```

```python
from konlpy.tag import Komoran
import pandas as pd

print('step1')
tokenizer = Komoran()

print('step2-1')
Q = person_1

print('step2-2')
A = person_2

print('step3-1')
sptd_Q = [tokenizer.morphs(Q[i]) for i in range(len(Q))]

print('step3-2')
sptd_A = [tokenizer.morphs(A[i]) for i in range(len(A))]

print('step4')
sptd_Q

print('step5')
sptd_A
```

```python
speech_data_df = pd.DataFrame({'Input': sptd_Q, 'Output': sptd_A})
speech_data_df.to_csv('./data/speech_data.txt', index = False, header = False, sep = '\t')
```
